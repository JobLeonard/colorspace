```python
from pytsp import atsp_tsp, run, dumps_matrix

matrix_sym = atsp_tsp(luv_distance_append, strategy="avg")

outf = "./luv_route.tsp"

with open(outf, 'w') as dest:
    dest.write(dumps_matrix(matrix_sym, name="LUV space Route"))

    tour = run(outf, start=0, solver="LKH")
tour
```
