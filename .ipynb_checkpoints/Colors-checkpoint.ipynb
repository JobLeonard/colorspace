```python
from pytsp import atsp_tsp, run, dumps_matrix

matrix = [
    [ 0,    2910, 4693 ],
    [ 2903, 0,    5839 ],
    [ 4695, 5745, 0    ]]
```

```python
matrix_sym = atsp_tsp(matrix, strategy="avg")
```

```python
outf = "./myroute.tsp"
with open(outf, 'w') as dest:
    dest.write(dumps_matrix(matrix_sym, name="My Route"))
```

```python
tour = run(outf, start=0, solver="LKH")
```
